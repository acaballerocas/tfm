## Uso de ExtendedGridSearchCV

En primer lugar, definimos los parÃ¡metros y estrategias que queremos considerar. Como solo es un ejemplo, en este caso no vamos a dividir la muestra en entrenamiento y test, usamos el dataset completo.

```python
# Importamos las funciones que vamos a utilizar
# Estrategias de imbalance:
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
# Modelo
from sklearn.linear_model import LogisticRegression
```

```python
# Definimos los diccionarios y listas de tuplas necesarios:

# HiperparÃ¡metros del modelo
parametros_modelo = {
    'penalty': [None, 'l2'],
    'C': [0.5, 1]
}

# Lista de estrategias para imbalance
estrategias = [
    ('ninguna', None),
    ('SMOTE', SMOTE(random_state=42)),
    ('ADASYN', ADASYN(random_state=42))
]

# HiperparÃ¡metros de las estrategias
parametros_estrategias = {
    'SMOTE__sampling_strategy': [0.3, 0.5, 1],
    'ADASYN__n_neighbors': [3,5]
}
```

Hacemos ExtendedGridSearchCV:

```python
from ExtendedModelSelection import ExtendedGridSearchCV

ExtGridSearch = ExtendedGridSearchCV(
    estimator=LogisticRegression(),
    param_grid=parametros_modelo,
    samplers=estrategias,
    sampler_params=parametros_estrategias,
    cv=3,
    scoring='average_precision'
)
```

```python
MejorModelo, MejoresParam, MejorScore = ExtGridSearch.fit(X,y)
```

En este caso, el mejor modelo consiste en:
- No utilizar ninguna estrategia de sobre/submuestreo
- Utilizar el modelo de regresiÃ³n logÃ­stica con C=0.5 y sin penalizacion.
